### Compose map with chart per province

Corresponds to figure 2 of manuscript

```python
from PIL import Image

# set input and output paths
input_folder = "../images/hist_provinces/"
output_file = "../images/hist_provinces/final_composite.png"

# -----------------------------
# Load background image
# -----------------------------
background = Image.open(input_folder
                        +"background_map.png").convert("RGBA")

# -----------------------------
# Define foreground images and positions
# coordinates = top-left corner
# -----------------------------
"""Algarve.png
Alto_Alentejo.png
background_map.png
Baixo_Alentejo.png
Beira_Alta.png
Beira_Baixa.png
Beira_Litoral.png
Douro_Litoral.png
Estremadura.png
Minho.png
Ribatejo.png
Trás-os-Montes_e_Alto_Douro.png
"""

placements_left = ["Minho_nolabel.png", "Douro_Litoral_nolabel.png", "Beira_Litoral_nolabel.png", "Ribatejo_nolabel.png", "Estremadura_nolabel.png", "Algarve_nolabel.png"]
placements_right = ["Trás-os-Montes_e_Alto_Douro_nolabel.png", "Beira_Alta_nolabel.png", "Beira_Baixa_nolabel.png", "Alto_Alentejo_nolabel.png","Baixo_Alentejo_nolabel.png"]


# -----------------------------
# Paste images left
# -----------------------------

# x-coord and y-coord for initial placement of the first image
x = 280
y = 0

for file in placements_left:
    fg = Image.open(input_folder + file).convert("RGBA")

    # Optional resize
    fg = fg.resize((1200, 600))

    # Paste with transparency
    background.paste(fg, (x, y), fg)

    # Update y-coordinate for the next image
    y += 650

# -----------------------------
# Paste images right
# -----------------------------

# x-coord and y-coord for initial placement of the first image
x = 3600
y = 300 

for file in placements_right:
    fg = Image.open(input_folder + file).convert("RGBA")

    # Optional resize
    fg = fg.resize((1200, 600))

    # Paste with transparency
    background.paste(fg, (x, y), fg)

    # Update y-coordinate for the next image
    y += 650


# add whole chart
# fg = Image.open(input_folder + "whole_dataset.png").convert("RGBA")
    
# Optional resize
# fg = fg.resize((3200, 1400))

 # Paste with transparency
# background.paste(fg, (1600, 3900), fg)

# -----------------------------
# Save result
# -----------------------------

# resize to A4 dimensions at 600 dpi (2480 x 3508 pixels)
# background = background.resize((2480, 3508))
background.save("../images/hist_provinces/final_composite.png", dpi=(600, 600))
print("Saved final_composite.png")
```
