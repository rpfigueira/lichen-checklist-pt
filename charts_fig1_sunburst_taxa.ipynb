```python
import pandas as pd
import os
import numpy as np

df = pd.read_csv('../dados_raw/checklist-species-29Abr2026.csv')

# Filter for taxonRank not genus
df = df[df['taxonRank'].str.lower() != 'genus'].reset_index(drop=True)
df.shape
df.head()
```

```python
# # ------------------------------------------------------------------
# 1.  Replace the “blank” and “Incertae sedis” values with NaN
#     so that they are automatically ignored by nunique(dropna=True)
# ------------------------------------------------------------------
higher_ranks = ['phylum', 'class', 'order', 'family', 'genus']


ignore_vals = ['', 'Incertae sedis']

# Keep the original data intact – just work on a copy
df_species_clean  = df.replace(ignore_vals, np.nan)
```

```python
df_species_clean.head()
```

### número de nomes

```python
# count unique taxonID (excluding NaN)
unique_taxonid_count = df_species_clean['taxonID'].nunique(dropna=True)
unique_taxonid_count
```

### numero de taxa aceites

```python
# count unique taxonID where status is 'accepted'
accepted_taxa_unique_count = df_species_clean.loc[df['taxonomicStatus'].fillna('').str.lower() == 'accepted', 'taxonID'].nunique()


accepted_taxa_unique_count
```

```python
# count occurrences of each taxonRank - total
df_taxonrank_counts = df_species_clean['taxonRank'].value_counts().reset_index()
df_taxonrank_counts.columns = ['taxonRank', 'count']
df_taxonrank_counts
```

```python
# filter for accepted taxa only
df_accepted = df_species_clean[df_species_clean['taxonomicStatus'].str.lower() == 'accepted'].reset_index(drop=True)
df_accepted.head()
df_accepted.shape
```

```python
# count occurrences of each taxonRank - accepted taxa only
df_taxonrank_counts = df_accepted['taxonRank'].value_counts().reset_index()
df_taxonrank_counts.columns = ['taxonRank', 'count']
df_taxonrank_counts
```

```python
# count number of families and genus pf accepted taxa
print(f"Number of families in accepted taxa: {df_accepted['family'].nunique()}")
print(f"Number of genera in accepted taxa: {df_accepted['genus'].nunique()}")
```

```python
# count occurrences of each taxonRank - accepted taxa only
df_taxonrank_counts = df_accepted['taxonRank'].value_counts().reset_index()
df_taxonrank_counts.columns = ['taxonRank', 'count']
df_taxonrank_counts
```

```python
# count number of families and genus pf accepted taxa without 'nom. nudum'
print(f"Number of families in accepted taxa: {df_accepted['family'].nunique()}")
print(f"Number of genera in accepted taxa: {df_accepted['genus'].nunique()}")
```

```python
# count number of genus that contain only one species
genus_species_counts = df_accepted.groupby('genus')['taxonID'].nunique().reset_index()
genus_with_one_species = genus_species_counts[genus_species_counts['taxonID'] == 1]
print(f"Number of genus that contain only one species: {genus_with_one_species.shape[0]}")
```

```python
# filter columns for taxonomy tree sunburst plot
df_taxonomy = df_accepted[['kingdom', 'phylum', 'class', 'order', 'family', 'genus', 'scientificName']].copy()
df_taxonomy.head()
df_taxonomy = df_taxonomy.drop_duplicates()
df_taxonomy = df_taxonomy.reset_index(drop=True)
df_taxonomy.head()
df_taxonomy.shape
```

```python
df_taxonomy = df_taxonomy.fillna("inc. sed.")
df_taxonomy = df_taxonomy[df_taxonomy['kingdom'] == "Fungi"].reset_index(drop=True)
```

```python
df_taxonomy.isna().any()
```

```python
df_taxonomy['sps_count'] = df_taxonomy.groupby('genus')['scientificName'].transform('nunique')
```

```python
import plotly.express as px


# columns for sunburst plot: family, genus, scientificName
levels = ['class', 'order','family', ]
missing_cols = [c for c in levels if c not in df_accepted.columns]
if missing_cols:
    raise KeyError(f"Missing expected column(s) in df_parts: {missing_cols}")

# prepare aggregated counts, replace missing values with "Unknown"
df_h = df_accepted[levels].fillna("Unknown")
counts = df_h.groupby(levels, as_index=False).size().rename(columns={'size': 'count'})

# create sunburst
fig = px.sunburst(
    counts,
    path=levels,
    values='count',
    color='count',
    color_continuous_scale='Viridis',
)
fig.update_layout(margin=dict(t=40, l=0, r=40, b=0))
fig.update_layout(font_size=36)
fig.update_layout(coloraxis_colorbar_title='Number<br>of species')
fig.show()
```

```python
# save figure as high-res png
out = os.path.join("../images", "taxonomic_sunburst_highres.png")
fig.write_image(out, format="png", width=2500, height=2000, scale=1)
```

```python
fig = px.treemap(
    df_taxonomy,
    path=['kingdom', 'phylum', 'class', 'order', 'family', 'genus'],
    maxdepth=3,
    color='family'
)
fig.update_layout(margin=dict(t=0, l=0, r=0, b=0))
fig.update_layout(width=600, height=600)
fig.show()
```
